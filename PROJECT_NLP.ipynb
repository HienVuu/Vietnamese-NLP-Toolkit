##Tách từ bằng Longest Matching




```python
#Sử dụng Regular Expression tách các âm tiết (syllables) trong mỗi câu, lưu thành 1 list
def syllablize(sentence):
    word = '\w+'
    non_word = '[^\w\s]'
    digits = '\d+([\.,_]\d+)+'

    patterns = []
    patterns.extend([word, non_word,digits])
    patterns = f"({'|'.join(patterns)})"
    sentence = ud.normalize('NFC', sentence)
    tokens = re.findall(patterns, sentence, re.UNICODE)
    return [token[0] for token in tokens]

#Load các file n_grams đã xây dựng
def load_n_grams(path):
    with open(path, encoding='utf8') as f:
      words = f.read().split('\n')
    return words

#Thuật toán Longest Matching
def longest_matching(sentence, bi_grams, tri_grams, quadri_grams, penta_grams):
    syllables = syllablize(sentence)
    syl_len = len(syllables)

    curr_id = 0
    word_list = []
    done = False

    while (curr_id < syl_len) and (not done):
      curr_word = syllables[curr_id]
      if curr_id >= syl_len - 1:
        word_list.append(curr_word)
        done = True
      else:
        second_word = syllables[curr_id + 1]
        pair_word = ' '.join([curr_word, second_word])
        if curr_id >= syl_len - 2:
          if (pair_word in bi_grams) or (pair_word.lower() in bi_grams):
            word_list.append('_'.join([curr_word, second_word]))
            curr_id += 2
          else:
            word_list.append(curr_word)
            curr_id += 1
        else:
          third_word = syllables[curr_id + 2]
          triple_word = ' '.join([pair_word, third_word])
          if curr_id >= syl_len - 3:
            if (triple_word in tri_grams) or (triple_word.lower() in tri_grams):
              word_list.append('_'.join([curr_word, second_word, third_word]))
              curr_id += 3
            elif (pair_word in bi_grams) or (pair_word.lower() in bi_grams):
              word_list.append('_'.join([curr_word, second_word]))
              curr_id += 2
            else:
              word_list.append(curr_word)
              curr_id += 1
          else:
            fourth_word = syllables[curr_id + 3]
            quadra_word = ' '.join([triple_word, fourth_word])
            if curr_id >= syl_len - 4:
              if (quadra_word in quadri_grams) or (quadra_word.lower() in quadri_grams):
                word_list.append('_'.join([curr_word, second_word, third_word, fourth_word]))
                curr_id += 4
              elif (triple_word in tri_grams) or (triple_word.lower() in tri_grams):
                word_list.append('_'.join([curr_word, second_word, third_word]))
                curr_id += 3
              elif (pair_word in bi_grams) or (pair_word.lower() in bi_grams):
                word_list.append('_'.join([curr_word, second_word]))
                curr_id += 2
              else:
                word_list.append(curr_word)
                curr_id += 1
            else:
              fifth_word = syllables[curr_id + 4]
              penta_word = ' '.join([quadra_word, fifth_word])
              if (penta_word in penta_grams) or (penta_word.lower() in penta_grams):
                word_list.append('_'.join([curr_word, second_word, third_word, fourth_word, fifth_word]))
                curr_id += 5
              elif (quadra_word in quadri_grams) or (quadra_word.lower() in quadri_grams):
                word_list.append('_'.join([curr_word, second_word, third_word, fourth_word]))
                curr_id += 4
              elif (triple_word in tri_grams) or (triple_word.lower() in tri_grams):
                word_list.append('_'.join([curr_word, second_word, third_word]))
                curr_id += 3
              elif (pair_word in bi_grams) or (pair_word.lower() in bi_grams):
                word_list.append('_'.join([curr_word, second_word]))
                curr_id += 2
              else:
                word_list.append(curr_word)
                curr_id += 1

    return word_list
```

```python
#Hàm đếm số từ đúng
def count_correct_words(pred, source, n_grams=5):
    pred_words = pred.split() #tách từng cụm trong string
    source_words = source.split()

    total_true, tp = 0, 0
    total_errors, fp = 0, 0
#xóa tất cả từ tách khác nhau
    idx = 0
    while idx < len(pred_words): #duyệt qua từng cụm trong string longest hoặc vncore
        if pred_words[idx] not in source_words[idx:(idx + n_grams)]: #nếu cụm từ trong string1 không nằm trong vòng 5 từ trong string thủ công
            if '_' in pred_words[idx]: fp += 1 #tách từ sai
            del pred_words[idx] #xóa phần tử
            total_errors += 1 #lỗi
        else: idx += 1

    idx = 0
    while idx < len(source_words): #duyệt qua string thủ công
        if source_words[idx] not in pred_words[idx:(idx + n_grams)]:
            del source_words[idx]
        else: idx += 1
# các từ tách đúng còn lại
    if len(pred_words) < len(source_words): words = pred_words
    else: words = source_words

    for idx in range (len(words)):
        if pred_words[idx] == source_words[idx]:
            if '_' in pred_words[idx]: tp += 1
            total_true += 1

    return total_true, total_errors, tp, fp
```

```python
#Hàm đánh giá tách
def tokenize_evaluation(pred, source, n_grams=5):
    total_true = 0
    total_errors = 0
    total_words = 0

    pred_tp = 0
    pred_fp = 0

    for pred_sentence, source_sentence in zip(pred, source):
        total_words += len(source_sentence.split()) #tach ra tung cụm từ một trong string
        if pred_sentence != source_sentence: #string khác string
            true, error, tp, fp = count_correct_words(pred_sentence, source_sentence, n_grams)
            total_true += true
            total_errors += error
            pred_tp += tp
            pred_fp += fp
        else: #nếu 2 string giống nhau
            for word in source_sentence.split():
                if '_' in word: pred_tp += 1 #đếm tự được tách
                total_true += 1
    return {
        'Accuracy': total_true / total_words,
        'Precision': pred_tp / (pred_tp + pred_fp),
        'Recall': pred_tp / count_manual_compounds,
        'True Positive': pred_tp,
        'False Positive': pred_fp,
        'Total True': total_true,
        'Total Errors': total_errors,
        'Total Words': total_words
    }
```

##Các hàm tạo đặc trưng

```python
def refix(word):
  if "_" in word:
    i = word.index('_')
    return word[0:i]
  else: return word
```

```python
def postfi(word):
  if "_" in word:
    i = word.rfind('_')
    return word[i+1:len(word)]
  else: return word
```

```python
def reduplicate(word):
  word = word.lower()
  if word.count("_") == 1: #từ 2 âm tiết giống i nhau
    i = word.index('_')
    if word[0:i] == word[i+1:len(word)]:
      return True
  return False
```

```python
def ambigous(word):
  word = word.lower()
  if word in dictionary:
    return dictionary[word]
  else: return 0
```

```python
def form_of_word(word):
  form = ""
  for i in range(len(word)):
    if word[i].isupper() == True:
      form += "A"
    elif word[i] == "_":
      form += "_"
    elif i == 0 or word[i-1] == "_":
      form += "a"
  return form
```

##Trích đặc trưng

```python
def feature_extraction(sentence):
  features = []
  for word in sentence:
    i = sentence.index(word)
    feature = {
        'word': word,
        'is_first_word': i == 0,
        'is_last_word': i == len(sentence) - 1,
        'pre_word': sentence[i-1] if i > 0 else '<START>',
        'next_word': sentence[i+1] if i < len(sentence)-1 else '<END>',
        'prefix': refix(word),
        'postfi': postfi(word),
        'reduplication': reduplicate(word),
        'ambigous_tags': ambigous(word),
        'form': form_of_word(word)
    }
    features.append(feature)
  return features #một list các dict
```

```python
def extract_verb_phrases(words, tags):
    verb_phrases = []
    temp_vp = []  # Danh sách lưu các từ trong VP

    for i, (word, tag) in enumerate(zip(words, tags)):
        # Phần phụ trước động từ (Trạng từ, Trợ từ)
        if tag in {'R', 'T'}:  # Trạng từ(phó từ), trợ từ
            temp_vp.append(word)

        elif tag == 'V':  # Động từ trung tâm
            temp_vp.append(word)

            # Kiểm tra bổ ngữ sau động từ (Đối tượng, địa điểm, thời gian, tính từ, giới từ)
            while i + 1 < len(tags):
                next_tag = tags[i + 1]
                next_word = words[i + 1]
                i += 1

                # Kiểm tra nếu sau động từ là tính từ, giới từ, danh từ chỉ địa điểm
                if next_tag in {'R','A', 'E', 'N', 'Nc', 'Nu'}:
                    temp_vp.append(next_word)
                elif next_tag == 'V':  # Kiểm tra nếu gặp động từ kế tiếp
                    temp_vp.append(next_word)
                else:
                    break  # Nếu không phải, kết thúc cụm VP

            # Kiểm tra nếu là một cụm động từ hợp lệ (có ít nhất 2 từ)
            if len(temp_vp) > 1:
                verb_phrases.append(" ".join(temp_vp))
            temp_vp = []  # Reset sau khi lưu VP

        # Nếu gặp dấu câu hoặc từ không thuộc VP, kết thúc cụm VP
        elif tag in {'CH', 'C', 'Cc', 'Z'}:
            if len(temp_vp) > 1:  # VP phải có ít nhất 2 từ
                verb_phrases.append(" ".join(temp_vp))
            temp_vp = []  # Reset sau khi gặp dấu câu hoặc từ không thuộc VP

    # Kiểm tra phần còn lại của `temp_vp` nếu có
    if temp_vp and len(temp_vp) > 1:
        verb_phrases.append(" ".join(temp_vp))

    return verb_phrases

# Hàm Pos_tag đã chỉnh sửa
def Pos_tag(text):
    # Giả sử rằng bạn đã có các hàm SliptDoc, WordSeg, và feature_extraction
    text_slip = SliptDoc(text)
    # Tách từ trong từng câu thành các danh sách từ
    text_seg = WordSeg(text_slip)  # list các string
    seg = []
    for x in text_seg:
        seg += x
    # Phân tích cú pháp để lấy các đặc trưng từ
    feature = []
    for x in text_seg:
        feature += feature_extraction(x)
    # Biến chuyển đổi văn bản thành đặc trưng (features)
    text_vec = vectorizer.transform(feature)
    # Dự đoán nhãn POS cho từng từ
    predicted_tags = svm_model.predict(text_vec)
    # Áp dụng chunking để tìm các cụm động từ (VP)
    verb_phrases = extract_verb_phrases(seg, predicted_tags)
    return seg, predicted_tags, verb_phrases

# Ví dụ văn bản
Text = "Con ngựa đá con ngựa đá ."
text_seg, tag, verb_phrases = Pos_tag(Text)

# In kết quả phân tách
print("Danh sách từ:", text_seg)
print("Danh sách nhãn POS:", tag)

# In kết quả cụm động từ (VP)
print("Cụm động từ (VP):")
for vp in verb_phrases:
    print(vp)

# Hiển thị kết quả theo dạng tuple
result = zip(text_seg, tag)
print("Kết quả dạng tuple:", tuple(result))
```
